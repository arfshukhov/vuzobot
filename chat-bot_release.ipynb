```python
#здесь представлен словарь университетов в соответствтии с профилем, городом и рангом (подробнее о рангах ниже)
universitys: dict = {
    "Tехнологический профиль":
        {
            "Москва": {
                "1": "Российский университет транспорта\nhttps://www.miit.ru",
                "2": "Московский политехнический университет\nhttps://mospolytech.ru",
                "3": "Российский новый университет\nhttps://rosnou.ru",
                "4": "Московский государственный университет имени М.В. Ломоносова\nhttps://www.msu.ru"
            },
            "Санкт-Петербург": {
                "1": "Государственный университет флота имени адмирала С.О. Макарова\nhttps://gumrf.ru ",
                "2": "Санкт-Петербургский государственный морской технический университет\nhttps://www.smtu.ru",
                "3": "Санкт-Петербургский Гуманитарный университет профсоюзов\nhttps://www.gup.ru",
                "4": "СПБГУ - Санкт-Петербургский государственный университет\nhttps://spbu.ru"
            },
            "Екатеринбург": {
                "1": "Уральский государственный университет путей сообщения\nhttps://www.usurt.ru",
                "2": "Уральский федеральный университет имени Б.Н. Ельцина\nhttps://urfu.ru/ru/",
                "3": "Уральский государственный лесотехнический университет\nusfeu.ru",
                "4": "Уральский институт связи и информатики\nhttps://www.uisi.ru"
            },
            "Оренбург": {
                "1": "Оренбургский государственный университет\nhttp://www.osu.ru",
                "2": "Оренбургский институт путей сообщения\nhttps://www.samgups.ru",
                "3": "Оренбургский государственный педагогический университет\nospu.ru",
                "4": "Оренбургский филиал Поволжского университета телекоммуникаций и информатики\nhttps://of.psuti.ru"
            }
        },
    "Гуманитарный профиль": {
        "Москва": {
            "1": "Московский информационно-технологический университет\nhttps://masi.ru",
            "2": "Государственный университет по землеустройству\nhttps://www.guz.ru",
            "3": "Московский государственный гуманитарно-экономический университет\nhttps://mggeu.ru",
            "4": "Всероссийская академия внешней торговли\nhttps://www.vavt.ru"
        },
        "Санкт-Петербург": {
            "1": "Петербургский государственный университет Императора Александра I\nhttps://www.pgups.ru",
            "2": "Северо-Западный институт управления\nhttps://spb.ranepa.ru",
            "3": "Институт правоведения и предпринимательства\nhttps://www.иппспб.рф",
            "4": "Санкт-Петербургский Гуманитарный университет профсоюзов\nhttps://www.gup.ru"},
        "Екатеринбург": {
            "1": "Уральский федеральный университет имени Б.Н. Ельцина\nhttps://urfu.ru/ru/",
            "2": "Институт международных связей\nhttps://ims-ural.ru",
            "3": "Уральский институт управления\nhttps://ui.ranepa.ru",
            "4": "Уральский государственный юридический университет имени В.Ф. Яковлева\nusla.ru"
        },
        "Оренбург": {
            "1": "Оренбургский государственный аграрный университет\nhttps://orensau.ru",
            "2": "Оренбургский государственный педагогический университет\nospu.ru",
            "3": "Оренбургский институт имени О.Е. Кутафина (МГЮА)\nhttp://www.oimsla.edu.ru",
            "4": "Оренбургский государственный институт искусств им. Л. и М. Ростроповичей\nhttps://osiart.ru"
        }
        },
    "Естественно-научный профиль": {
        "Москва": {
            "1": "Российский государственный аграрный университет имени К.А. Тимирязева\nhttps://www.timacad.ru",
            "2": "МИРЭА – Российский технологический университет\nhttps://www.mirea.ru",
            "3": "Российский биотехнологический университет\nhttps://international.mgupp.ru",
            "4": "МГУ - Московский государственный университет имени М.В. Ломоносова\nhttps://www.msu.ru"},
        "Санкт-Петербург": {
            "1": "Санкт-Петербургский государственный лесотехнический университет\nhttps://spbftu.ru/public/",
            "2": "Северо-Западный государственный медицинский университет имени И.И. Мечникова\nhttps://szgmu.ru",
            "3": "Национальный исследовательский университет ИТМО\nitmo.ru",
            "4": "Институт медицинского образования имени В. А. Алмазова\nhttp://education.almazovcentre.ru"
        },
        "Екатеринбург": {
            "1": "Уральский федеральный университет имени Б.Н. Ельцина\nhttps://urfu.ru/ru/",
            "2": "Уральский государственный медицинский университет\nusma.ru",
            "3": "Уральский государственный лесотехнический университетn\nusfeu.ru",
            "4": "Уральский государственный медицинский университет\nusma.ru"
        },
        "Оренбург": {
            "1": "Оренбургский государственный аграрный университет\norensau.ru",
            "2": "Оренбургский государственный университет\nospu.ru",
            "3": "Оренбургский государственный медицинский университет\n",
            "4": "orgma.ru\n"
        }
        }
}


#список городов, используется в одной из функций
CITIES = ["Москва", "Санкт-Петербург", "Оренбург", "Екатеринбург"]


#стартовая функция
def start():
    print("""Привет! Я чат-бот и могу помочь тебе определиться с выбором ВУЗа. Пока что я 
могу работать только по нескольким городам.""")
    print('''Напишите в консоль "НАЧАТЬ!"\n''')
    # хендлер исключений, обрабатывает случай прерывания программы без ввода или с некорректным вводом. 
    # Все дальнейшие блоки try-except выполняют ту же функцию
    try:
        word = str(input())
        if word == "НАЧАТЬ!":
            select_profile()
        else:
            start()
    except KeyboardInterrupt:
        print("Непредвиденное поведение. Перезапустите программy")
    except Exception:
        print("Непредвиденное поведение. Перезапустите программу")
        
        
def select_profile():
    print("Отлично! Теперь давай определимся с твоим профилем")
    print("1 - Tехнологический профиль", "2 - Гуманитарный профиль", "3 - Естественно-научный профиль", sep="\n")
    profiles = {
        "1": "Tехнологический профиль",
        "2": "Гуманитарный профиль",
        "3": "Естественно-научный профиль"
    }
    try:
        answer = input()
        # соотносим введенное число соответствующего профиля со словарем и обрабатываем случаи, когда в качестве ответа введена неверная строка
        while answer not in ("1", "2", "3"):
            print("Такого варианта ответа нет, попробуй еще раз")
            answer = input()
        else:
            processing_profile(profiles.get(answer))
    except KeyboardInterrupt:
        print("Непредвиденное поведение. Перезапустите программy")
    except Exception:
        print("Непредвиденное поведение. Перезапустите программу")


def processing_profile(profile: str = None):
    print(f"Супер, выбран {profile}")
    print('Если все верно, введи "1", если что-то не так, ты можешь вернуться назад, введя "2"')
    try:
        answer = input()
        # пользователю предлагается свериться с данными, которые он ввел. Если он захочет изменить, то сможет вернуться к предыдущему шагу
        while answer != "1" and answer != "2":
            print("Такого ответа нет")
            answer = input()
        else:
            if answer == "1":
                processing_results(profile)
            elif answer == "2":
                select_profile()
    except KeyboardInterrupt:
        print("Непредвиденное поведение. Перезапустите программy")
    except Exception:
        print("Непредвиденное поведение. Перезапустите программу")
        
def processing_results(profile: str = None):
    #У пользователя запрашивается количество предметов и сумма результатов и проводится провека на сооветствие шаблону
    try:
        count_of_exam = int(input("Введи количество экзаменов (от 2 до 10): "))
        sum_of_results = int(input("Теперь введи сумму своих баллов (от 11 до 400): "))
    except KeyboardInterrupt:
        print("Непредвиденное поведение. Перезапустите программy")
        sys.exit()
    except Exception:
        print("Непредвиденное поведение. Перезапустите программу")
        sys.exit()
    if count_of_exam not in range(2, 11) or sum_of_results not in range(11, 401):
        print("Вы ввели недопустимое значение")
        processing_results(profile)
    else:
        rank = get_rank(count_of_exam, sum_of_results)
        select_city(profile, rank)
        
        
def get_rank(cnt: int, res: int) -> str:
    # производится условное разделение на ранг в соответствии со средним результатом ЕГЭ
    # разделение ОЧЕНЬ условное, как и подбор ВУЗов согласно нему
    # в будущем возможно будет подобрать более подходящие ВУЗы некоторым рангам
    med = res//cnt
    rank: str = "0"
    if 0 <= med <= 33:
        rank = "1"
    elif 34 <= med <= 55:
        rank = "2"
    elif 56 <= med <= 79:
        rank = "3"
    elif 80 <= med:
        rank = "4"
    return rank


def select_city(profile: str, rank: str):
    print("Теперь выбери город. Доступны следующие города")
    print("1 - Москва", "2 - Санкт-Петербург", "3 - Екатеринбург", "4 - Оренбург")
    answer = ""
    try:
        answer = input()
    except KeyboardInterrupt:
        print("Непредвиденное поведение. Перезапустите программy")
        sys.exit()
    except Exception:
        print("Непредвиденное поведение. Перезапустите программу")
        sys.exit()
    # определяем индекс
    match answer:
        case "1":
            city = "Москва"
        case "2":
            city = "Санкт-Петербург"
        case "3":
            city = "Оренбург"
        case "4":
            city = "Екатеринбург"
        case "НАЗАД!":
            start()
        case _:
            print("Такого варианта ответа нет")
            select_city(profile, rank)
    # определяем город через по индексу
    #константный список из начала программы
    city = CITIES[int(answer)-1]
    #вызывается функция get_university с параметрами, установленным ранее при выполнении программы
    university = get_university(profile, city, rank)
    final(university)
    

def get_university(profile: str, city: str, rank: str) -> str:
    # достаем из словаря университетов подходящий нам по данным
    return universitys[profile][city][rank]


def final(university: str):
    print("Для вас подобран университет:")
    print(university)
    print('Если вы хотите посмотреть другие варианты, то введите "ЗАНОВО"')
    # Пользователь может завершить выполнение программы, либо вернуться к первому шагу
    try:
        answer = input()
        if answer == "ЗАНОВО":
            # сценарий, при котором пользователь возвращается к первому шагу
            start()
        else:
            # сценарий, при котором пользователь выходит из программы
            print("Пока-пока!")
            sys.exit()

    except KeyboardInterrupt:
        sys.exit()
    except Exception:
        sys.exit()


#точка входа выполнение программы        
if __name__ == "__main__":
    #вызов стартовой функции
    start()
```
